```python
# Define the list of product categories to exclude
excluded_products = ['Distribution Losses', 'Electricity Supplied', 'Electricity Trade', 'Final Consumption', 'Low Carbon', 
                    'Net electricity production', 'Not specified', 'Other combustible non-renewables', 'Other renewables', 
                    'Other renewables aggregated', 'Others', 'Total combustible fuels', 'Total exports', 'Total imports'
                    'Used for pumped storage']  

# Exclude the specified product categories from the DataFrame
df_filter = df_filt[~df_filt['PRODUCT'].isin(excluded_products)]

# Group the filtered data by year and count the frequency of each product
product_counts = df_filter.groupby('YEAR')['PRODUCT'].value_counts().unstack(fill_value=0)

# # Create a custom color cycle for the lines
# color_cycle = plt.cm.Set1.colors  # You can choose any colormap you like

# Create a line chart using Matplotlib
plt.style.use('classic')

# # Set the color cycle for the lines
# plt.gca().set_prop_cycle(color=color_cycle)

product_counts.plot(kind='line', marker='o', figsize=(10, 6))

plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Frequency of Products (2010-2020)')

plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# Group the filtered data by year and product, summing the values
product_counts = df_filter.groupby(['YEAR', 'PRODUCT'])['VALUE'].sum().unstack(fill_value=0)

# Create a bar chart using Matplotlib
plt.figure(figsize=(10, 6))

# Define custom colors for the bars (optional)
bar_colors = ['blue', 'green', 'red', 'purple', 'orange']  # Add more colors as needed

# Initialize a color index for bars (optional)
color_index = 0

# Iterate over unique product categories
for product_category in df_filter['PRODUCT'].unique():
    # Bar chart for the current product category
    plt.bar(
        product_counts.index,
        product_counts[product_category],
        label=product_category,
        color=bar_colors[color_index] if bar_colors else None,
    )
    
    # Increment the color index (optional)
    color_index = (color_index + 1) % len(bar_colors)

plt.xlabel('Year')
plt.ylabel('Total Value')
plt.title('Bar Chart of Products (2010-2020)')
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
